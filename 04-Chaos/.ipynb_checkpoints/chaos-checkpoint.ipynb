```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
### Define Variables for Function ###

a = 0.25
b = 4
F = 8
G = 1

#Time (days)

Tend = 120
T0 = 0
Tnorm = 5
DelT = 1/6

#Time Normalized

TE_Norm = 24
TS_Norm = 0
DelT_Norm = 1/30

#Time Steps

T_Steps = np.arange(T0,Tend,DelT)

#Other Initial Values

X0 = 2.5
X02 = 2.49
Y0 = 1
Z0 = 0

print(T_Steps)
```

```python
### Calculate Functions at T0 and Tend ###

#At First Time Step

X_t = X0 + DelT_Norm*(-(Y0)**2 - (Z0)**2 - a*(X0) + a*F)
Y_t = Y0 + DelT_Norm*(X0*Y0 - b*X0*Z0 - Y0 + G)
Z_t = Z0 + DelT_Norm*(b*X0*Y0 + X0*Z0 - Z0)

#For Subsequent Time Steps

Xtot = X0 + X_t*(T_Steps)
Xtot2 = X02 + X_t*(T_Steps)
Ytot = Y0 + Y_t*(T_Steps)
Ztot = Z0 + Z_t*(T_Steps)
```

```python
#Now Calculate Final Values

Xf=np.zeros(T_Steps.shape)
Xf2=np.zeros(T_Steps.shape)
Yf=np.zeros(T_Steps.shape)
Yf2=np.zeros(T_Steps.shape)
Zf=np.zeros(T_Steps.shape)
Zf2=np.zeros(T_Steps.shape)
Xf[0] = X0
Xf2[0] = X02
Yf[0] = Y0
Yf2[0] = Y0
Zf[0] = Z0
Zf2[0] = Z0

#For X0=2.5
for i in range(1,len(Xf)):
    Xf[i] = Xf[i-1] + DelT_Norm*(-(Yf[i-1])**2 - (Zf[i-1])**2 - a*(Xf[i-1]) + a*F)
    Yf[i] = Yf[i-1] + DelT_Norm*(Xf[i-1]*Yf[i-1] - b*Xf[i-1]*Zf[i-1] - Yf[i-1] + G)
    Zf[i] = Zf[i-1] + DelT_Norm*(b*Xf[i-1]*Yf[i-1] + Xf[i-1]*Zf[i-1] - Zf[i-1])

#For X0=2.49
for j in range(1,len(Xf)):
    Xf2[j] = Xf2[j-1] + DelT_Norm*(-(Yf[j-1])**2 - (Zf[j-1])**2 - a*(Xf2[j-1]) + a*F)
    Yf2[j] = Yf2[j-1] + DelT_Norm*(Xf2[j-1]*Yf2[j-1] - b*Xf2[i-1]*Zf[j-1] - Yf2[j-1] + G)
    Zf2[j] = Zf2[j-1] + DelT_Norm*(b*Xf2[j-1]*Yf2[j-1] + Xf2[i-1]*Zf2[j-1] - Zf2[j-1])
```

```python
#X plot
plt.plot(Xf)
plt.plot(Xf2)

#Plot Labels
plt.title("Circumpolar Westerlies for X=2.5 and X=2.49")
plt.xlabel("Time")
plt.ylabel("Strength of Westerlies")

#Save Figure
plt.savefig("Chaos_FigX")
```

```python
#Y Plot
plt.plot(Yf)
plt.plot(Yf2)

#Plot Labels
plt.title("Y Waves for X=2.5 and X=2.49")
plt.xlabel("Time")
plt.ylabel("Wave Amplitude")

#Save Figure
plt.savefig("Chaos_FigY")
```

```python
#Z Plot
plt.plot(Zf)
plt.plot(Zf2)

#Plot Labels
plt.title("Z Waves for X=2.5 and X=2.49")
plt.xlabel("Time")
plt.ylabel("Wave Amplitude")

#Save Figure
plt.savefig("Chaos_FigZ")
```
