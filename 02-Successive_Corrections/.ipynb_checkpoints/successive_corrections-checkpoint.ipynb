```python
### Read in observations ###
with open("C:/Users/Mark McGarry/Documents/ATSC_528/RAOBs_201903131200.txt") as file:
    data = file.read()

stn_name = [a.split(',')[0] for a in data.split('\n')]
print(stn_name)

data = np.genfromtxt(StringIO(data), delimiter=",")
data = data[:,1:]

latitude = data[:,0]
longitude = data[:,1]
Height = data[:,2]
plt.scatter(longitude, latitude)
```

```python
### Set up analysis map with a 22x28 rectangular grid of points ###
x = np.arange(22)*(1.27)+18.9
y = np.arange(28)*(1.27)-6.3

X,Y = np.meshgrid(x,y)
plt.scatter(X,Y)
```

```python
### convert obs lat/long to x,y ###

rho = 637100000 #radius of Earth in cm
m = 1/15000000
phi_O = 60*(np.pi / 180)
lam_O = -115*(np.pi / 180)

lat_rad = latitude*(np.pi / 180)
lon_rad = longitude*(np.pi / 180)

sigma = (1 + np.sin(phi_O)) / (1 + np.sin(lat_rad))

r = m * (rho * sigma * np.cos(lat_rad))
x_obs = r * (np.cos(lon_rad - lam_O))
y_obs = r * (np.sin(lon_rad - lam_O))

Ob_Lat = ((np.pi/2) - 2*np.arctan((np.sqrt((x_obs / m)**2+(y_obs / m)**2))/(rho*(1+np.sin(phi_O)))))*(180/np.pi)
Ob_Lon =  (np.arctan((y_obs/m)/(x_obs/m)))*(180/np.pi) - 115
Lat_Grid = ((np.pi/2) - 2*np.arctan((np.sqrt((X / m)**2+(Y / m)**2))/(rho*(1+np.sin(phi_O)))))*(180/np.pi)
Lon_Grid =  (np.arctan((Y/m)/(X/m)))*(180/np.pi) - 115

proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(8,8),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)

plt.scatter(Ob_Lon, Ob_Lat, transform = ccrs.PlateCarree())
plt.scatter(Lon_Grid, Lat_Grid, transform = ccrs.PlateCarree(), s = 5)
```

```python
### Define function for bilinear interpolation ###

def bilinear_int(x, y, z):

    #Create Local Grid#
    Xloc = x - np.floor(x)
    Yloc = y - np.floor(y)

    #Local Analysis Values#
    Z1 = z[int(np.floor(y)), int(np.floor(x))]
    Z2 = z[int(np.ceil(y)), int(np.floor(x))]
    Z3 = z[int(np.ceil(y)), int(np.ceil(x))]
    Z4 = z[int(np.floor(y)), int(np.ceil(x))]

    #Create Function#

    left = (Z2-Z1)*(Yloc)+Z1
    right = (Z3-Z4)*(Yloc)+Z4
    Pf = (right-left)*(Xloc)+left
    #P1 = (1 - Xloc)*(1 - Yloc)*(Z1)
    #P2 = (Xloc - 0)*(1 - Yloc)*(Z2)
    #P3 = (Xloc - 0)*(Yloc - 0)*(Z3)
    #P4 = (1 - Xloc)*(Yloc - 0)*(Z4)
    #Pf = ((P1 + P2) + (P3 + P4))
    
    return Pf

#Define Cressman Function#

def Cressman(dmin, dik, Rat):
    ROI = Rat*dmin
    if dik > ROI:
        w = 0
    else:
        w = (ROI**2 - dik**2)/(ROI**2 + dik**2)
    return w
```

```python
### Create function for bilinear interpolation ###

#Define Indices:
Xind = (x_obs - 18.9)/1.27
Yind = (y_obs + 6.3)/1.27
print(Xind, Yind)

Xtl = np.floor(Xind)
Xbl = np.floor(Xind)
Xtr = np.ceil(Xind)
Xbr = np.ceil(Xind)

Ytl = np.ceil(Yind)
Ybl = np.floor(Yind)
Ytr = np.ceil(Yind)
Ybr = np.floor(Yind)
```

```python
### Perform 500mb geopotential height analyses using a Cressman weighting Function###
#Use radii of influence 4, 2.5, 1.5 *dmin

dmin = np.zeros(135)
for i in range(135):
    xd = x_obs[i] - x_obs
    yd = y_obs[i] - y_obs
    td = np.sqrt(((xd)**2)+((yd**2)))
    dmin[i] = np.min(td[td>0.0001])
dmin = np.mean(dmin)
print(dmin)
print(td.shape)
```

```python
### First analysis, no successive corrections ###

Farray = np.zeros((len(y),len(x)))
for i in range(len(X)):
    for j in range(len(X[i])):
        xk = X[i,j] - x_obs
        yk = Y[i,j] - y_obs
        dik = np.sqrt((xk**2)+(yk**2))
        
        #Create List for Cressman Weighting#
        wlist = []
        for k in dik:
            wlist += [Cressman(dmin, dik=k, Rat=4)]
        wlist = np.array(wlist)
        A = np.sum(wlist*Height)/np.sum(wlist)
        Farray[i,j] = A
print(Farray)
```

```python
### 2nd Pass ###
bi_int_lst = []
Farray2 = np.zeros((len(y),len(x)))
for q in range(len(Xind)):
    if (0<Xind[q]<21) and (0<Yind[q]<27):
        B = bilinear_int(Xind[q], Yind[q], Farray)
    else:
        X_Loc = x_obs[q] - X
        Y_Loc = y_obs[q] - Y
        D_Loc = np.sqrt((X_Loc)**2 + (Y_Loc)**2)
        V_Good = np.where(D_Loc <= 4*dmin)
        W = np.zeros((X.shape))
        for r in V_Good[0]:
            for s in V_Good[1]:
                W[r,s] = Cressman(dmin, D_Loc[r,s], 4)
        B = np.sum(W*Farray)/np.sum(W)
    bi_int_lst += [B]
bi_int_array = np.array(bi_int_lst)
for a in range(len(X)):
    for f in range(len(X[a])):
        xk = X[a,f] - x_obs
        yk = Y[a,f] - y_obs
        dik = np.sqrt((xk**2)+(yk**2))
        
        #Create List for Cressman Weighting
        wlist = []
        for k in dik:
            wlist += [Cressman(dmin, dik=k, Rat=2.5)]
        wlist = np.array(wlist)
        Farray2[a,f] = Farray[a,f] + (np.nansum(wlist*(Height - bi_int_array))/np.sum(wlist))
print(Farray2)
```

```python
### 3rd Pass ###
bi_int_lst2 = []
Farray3 = np.zeros((len(y),len(x)))
for x in range(len(Xind)):
    if (0<Xind[x]<21) and (0<Yind[x]<27):
        B = bilinear_int(Xind[x], Yind[x], Farray2)
    else:
        X_Loc = x_obs[x] - X
        Y_Loc = y_obs[x] - Y
        D_Loc = np.sqrt((X_Loc)**2 + (Y_Loc)**2)
        V_Good = np.where(D_Loc <= 2.5*dmin)
        W = np.zeros((X.shape))
        for r in V_Good[0]:
            for s in V_Good[1]:
                W[r,s] = Cressman(dmin, D_Loc[r,s], 2.5)
        B = np.sum(W*Farray2)/np.sum(W)
    bi_int_lst2 += [B]
bi_int_array2 = np.array(bi_int_lst2)
for a in range(len(X)):
    for f in range(len(X[a])):
        xk = X[a,f] - x_obs
        yk = Y[a,f] - y_obs
        dik = np.sqrt((xk**2)+(yk**2))
        
        #Create List for Cressman Weighting
        wlist = []
        for k in dik:
            wlist += [Cressman(dmin, dik=k, Rat=1.5)]
        wlist = np.array(wlist)
        Farray3[a,f] = Farray2[a,f] + (np.nansum(wlist*(Height - bi_int_array2))/np.sum(wlist))
Farray3[np.isnan(Farray3)] = Farray2[np.isnan(Farray3)]        
```

```python
### Calculate Analysis Differences
Diff_1 = Farray2 - Farray
Diff_2 = Farray3 - Farray2
Diff_3 = Farray3 - Farray
```
